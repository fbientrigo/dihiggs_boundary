# CMS 13TeV LHC

## `t>H+b>cb` [1808.06575](https://arxiv.org/abs/1808.06575)


```python
df = pd.DataFrame(
    {
        "m": [90, 100, 110, 120, 130, 140, 150],
        "obs": [1.34e-2, 1.25e-2, 1.22e-2, 8.45e-3, 4.26e-3, 3.4e-3, 3.34e-3],
        "exp": [8.28e-3, 6.56e-3, 5.94e-3, 7.27e-3, 6.56e-3, 5.31e-3, 5.31e-3],
    }
)
limitFile = implementChannelLimit(
    "1808.06575",
    {"channels": [["brtHpb", "cb"]]},
    "https://www.hepdata.net/record/ins1688938",
    df,
    massResolution=resolution.cb["default"],
    constraints={"topDecayConsistency": ["brtHpb"]},
)
lim = HB.Limit(limitFile)
fig, ax = validateChannelLimit(lim)
```
