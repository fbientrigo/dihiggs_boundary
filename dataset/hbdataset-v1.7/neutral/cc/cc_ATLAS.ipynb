# ATLAS 13 TeV

## ZH > cc (h125) [1802.04329](https://arxiv.org/abs/1802.04329)

```python
df = pd.DataFrame({"m": [125], "obs": [2.7], "exp": [3.9]})


limitFile = implementChannelLimit(
    "1654814",
    {"channels": [["HZ", "cc"]]},
    "https://arxiv.org/abs/1802.04329",
    df,
    massResolution={"relative": 0, "absolute": 40},  # +- 20 GeV
    constraints={
        "modelLike": "SMHiggs",
        "process": {
            "channels": [["none", "bb"], ["none", "cc"]]
        },  # ZH>bb is major background
    },
)
lim = HB.Limit(limitFile)
pred = Higgs.Predictions()

h = pred.addParticle(HP.NeutralScalar("h"))
h.setMass(125)
HP.effectiveCouplingInput(h, HP.smLikeEffCouplings)

h.setChannelRate("LHC13", "HZ", "cc", 1)
assert np.isclose(lim.apply(pred)[0].expRatio(), 1 / df.exp[0])
assert np.isclose(lim.apply(pred)[0].obsRatio(), 1 / df.obs[0])
```
