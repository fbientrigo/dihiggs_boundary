# Example for computing HH production cross section

```python
import Higgs.predictions as HP
import Higgs.bounds as HB

pred = HP.Predictions() # create the model predictions
```

SM-like Higgs boson with SM-like couplings

```python
h = pred.addParticle(HP.BsmParticle("h", "neutral", "even"))
h.setDecayWidth("h", "h", 1)
h.setCxn("LHC13", "ggH", 1)
HP.effectiveCouplingInput(h, HP.smLikeEffCouplings)
```

cross section prediction

```python
xsecres =h.cxn('LHC13', 'pair')
print(xsecres)
```
